2) Replace missing values in Min.Price and Max.Price columns with their respective mean.

```python
min_avg = df["Min.Price"].mean()
max_avg = df["Max.Price"].mean()

filled_df = df
filled_df["Min.Price"] = df["Min.Price"].fillna(min_avg)
filled_df["Max.Price"] = df["Max.Price"].fillna(max_avg)

filled_df.tail(10)
```

3) How to get the rows of a dataframe with row sum > 100?

```python
rowsum_df = df.select_dtypes(include="number")
rowsum_df = rowsum_df.sum(axis=1)
mask = rowsum_df > 100
mask.value_counts()

#Looks like every row adds together over 100
```
