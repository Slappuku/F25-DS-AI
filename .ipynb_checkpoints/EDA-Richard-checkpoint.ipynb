```python
monthly = sp500.resample('M').mean(numeric_only=True)
smoothed = monthly.rolling(3).mean()  # 3-month moving average

plt.figure(figsize=(12,6))
plt.plot(smoothed.index, smoothed['GDP Growth (%)'], label='GDP (3-MA)')
plt.plot(smoothed.index, smoothed['Inflation Rate (%)'], label='Inflation (3-MA)')
plt.plot(smoothed.index, smoothed['Unemployment Rate (%)'], label='Unemployment (3-MA)')
plt.title('Smoothed Monthly Economic Indicators')
plt.legend()
plt.show()
```

```python
corrs = {}
for lag in range(-6, 7):  # -6 to +6 months
    corrs[lag] = smoothed['GDP Growth (%)'].corr(smoothed['Unemployment Rate (%)'].shift(lag))

plt.figure(figsize=(10,5))
plt.plot(corrs.keys(), corrs.values(), marker='o')
plt.title('Correlation between GDP Growth and Unemployment at Different Lags')
plt.xlabel('Lag (months)')
plt.ylabel('Correlation')
plt.grid(True)
plt.show()
```

```python
# Convert to Series for easy handling
corr_series = pd.Series(corrs)

# Find lags of maximum and minimum correlation
max_lag = corr_series.idxmax()
min_lag = corr_series.idxmin()

plt.figure(figsize=(10,6))
plt.plot(corr_series.index, corr_series.values, marker='o', linestyle='-')
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='black', linestyle=':', label='No lag')

# Highlight peaks
plt.scatter(max_lag, corr_series[max_lag], color='green', s=100, label=f'Max Corr ({max_lag:+}m) = {corr_series[max_lag]:.2f}')
plt.scatter(min_lag, corr_series[min_lag], color='red', s=100, label=f'Min Corr ({min_lag:+}m) = {corr_series[min_lag]:.2f}')

plt.title('Lead/Lag Correlation between GDP Growth and Unemployment')
plt.xlabel('Lag (months)')
plt.ylabel('Correlation')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))

# Original GDP growth
plt.plot(smoothed.index, smoothed['GDP Growth (%)'], label='GDP Growth (3-MA)', color='tab:blue')

# Unemployment shifted forward by 1 month to reflect -1 lag correlation
plt.plot(smoothed.index, smoothed['Unemployment Rate (%)'].shift(1),
         label='Unemployment (shifted +1 month)', color='tab:red', linestyle='--')

plt.title('Unemployment Leading GDP Growth by 1 Month')
plt.xlabel('Date')
plt.ylabel('Percent (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.scatter(smoothed['GDP Growth (%)'], smoothed['Unemployment Rate (%)'], alpha=0.6)
plt.title('GDP Growth vs Unemployment')
plt.xlabel('GDP Growth (%)')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.show()

print("Correlation:", smoothed['GDP Growth (%)'].corr(smoothed['Unemployment Rate (%)']))
```

```python
mean_infl = smoothed['Inflation Rate (%)'].mean()
high = smoothed[smoothed['Inflation Rate (%)'] > mean_infl]
low = smoothed[smoothed['Inflation Rate (%)'] <= mean_infl]

print("Average GDP Growth (high inflation):", high['GDP Growth (%)'].mean())
print("Average GDP Growth (low inflation):", low['GDP Growth (%)'].mean())
print("Average Unemployment (high inflation):", high['Unemployment Rate (%)'].mean())
print("Average Unemployment (low inflation):", low['Unemployment Rate (%)'].mean())
```

```python
window = 12  # 12 months = 1 year
rolling_corr = smoothed['GDP Growth (%)'].rolling(window).corr(smoothed['Unemployment Rate (%)'])

plt.figure(figsize=(10,5))
plt.plot(rolling_corr, label='Rolling Correlation (1Y)')
plt.axhline(0, color='gray', linestyle='--')
plt.title('Rolling Correlation: GDP vs Unemployment')
plt.legend()
plt.grid(True)
plt.show()
```

a –0.6 correlation means that when the economy was expanding (higher GDP growth), unemployment was falling, and when growth slowed, unemployment was rising
